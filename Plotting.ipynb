```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#import dataset
data=pd.read_csv("Iris.csv")
```

```python
data.head()
```

# Matplotlib.pyplot

## Sactter plot

```python
data.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='red')
plt.show()
```

## Bar plot

```python
data1=data.head(10)
```

```python
data1.plot(kind='bar',x='SepalLengthCm',y='SepalWidthCm')
```

## line plot

```python
# gca stands for 'get current axis'
ax = plt.gca()

data.plot(kind='line',x='SepalLengthCm',y='PetalLengthCm',ax=ax)
data.plot(kind='line',x='SepalLengthCm',y='PetalWidthCm', color='red', ax=ax)

plt.show()
```

```python
data[['SepalLengthCm']].plot(kind='hist',bins=[0,2,4,6,8,10],rwidth=0.8)
plt.show()
```

## stacked bar

```python
data2=data[["SepalLengthCm","PetalLengthCm","Species"]]
data2.head()
```

```python
data2=data2.groupby(["Species"]).sum()
data2
```

```python
data2.plot(kind='bar', stacked=True)
plt.title("stacked bar graph")
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 1, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(2, 1, 2) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: Box plot
data2.plot(kind='bar', figsize=(5, 5), logy=True, ax=ax0) # add to subplot 1
ax0.set_title('subplots')
ax0.set_xlabel('x')
ax0.set_ylabel('y')

# Subplot 2: Line plot
data2.plot(kind='bar', figsize=(5, 5), logy=True,stacked=True, ax=ax1) # add to subplot 2
ax1.set_title ('')
ax1.set_ylabel('y')
ax1.set_xlabel('')

plt.show()
```

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: Box plot
data2.plot(kind='bar', figsize=(20,5), logy=True, ax=ax0) # add to subplot 1
ax0.set_title('subplots')
ax0.set_xlabel('x')
ax0.set_ylabel('y')

# Subplot 2: Line plot
data2.plot(kind='bar', figsize=(20,5), logy=True,stacked=True, ax=ax1) # add to subplot 2
ax1.set_title ('')
ax1.set_ylabel('y')
ax1.set_xlabel('')

plt.show()
```
